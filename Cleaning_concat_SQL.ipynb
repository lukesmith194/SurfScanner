```python
nazare['Spot'] = list_nazare
```

```python
def apply_condi_naz(string):
    if int(string[3:5]) in ([10,11,3]) :
        return "Epic"
    elif int(string[3:5]) in([9,12,1,2]):
        return "Good"
    elif int(string[3:5]) in ([4]):
        return "Mediocre"
    else:
        return "Not recommended"
```

```python
nazare['Conditions'] = nazare.date.apply(apply_condi_naz)
```

### Concat dataframes into one

```python
full = pd.concat([full,tauro], ignore_index=True)
```

```python
full.drop(["mph","prob", "wind_mph"], axis = 1, inplace = True)
```

```python
full = full[['Spot', 'date', 'time', 'range', "height","swell_dir", "swell_angle","period","wind_kmh","wind_dir","wind","wind_angle","temp","Conditions"]]
```
