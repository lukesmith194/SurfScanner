```python
def create_df_nazare(string):
    new1 = []
    final = []
    b = re.split(r"[A-Z]{1,}", string)
    new1.append(b)
    for i in b:
        final.append(i.split("/n"))
    #string = string.replace("�","E")
    #string = string.replace("SEBADO","SABADO")
    df = pd.DataFrame(final)
    df = df[0].str.split(expand=True)
    df.head()
    topdf = clean_top(df)
    bottomdf = clean_bottom(df)
    topdf.columns = ["date","time","range","size","period","wind_km","wind_mph","water_temp","prob","time2","range2","size2","period2","wind_km2","wind_mph2","water_temp2","prob2","time3","range3","size3","period3","wind_km3","wind_mph3","water_temp3","prob3"]
    bottomdf.columns = ["date","time","range","size","period","wind_km","wind_mph","water_temp","prob","time2","range2","size2","period2","wind_km2","wind_mph2","water_temp2","prob2","time3","range3","size3","period3","wind_km3","wind_mph3","water_temp3","prob3"]
    full_df = pd.concat([topdf, bottomdf],ignore_index=True)
    return full_df
```
