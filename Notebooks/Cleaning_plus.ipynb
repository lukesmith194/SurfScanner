### Concat 2019

```python
full_2019.dtypes
```

```python
full_2019.head()
```

```python
full_2019 = pd.concat([nazare ,pipeline , fronton ,box,puerto_escondido ,itacoatiara ,padang_padang,mosca,tauro,mundaka], ignore_index=True)
```

```python
full_2019.drop(["mph","prob", "wind_mph"], axis = 1, inplace = True)
```

```python
full_2019 = full_2019[['Spot', 'date', 'time', 'range', "height","swell_dir", "swell_angle","period","wind_kmh","wind_dir","wind","wind_angle","temp","Conditions"]]
```

```python
full_2019[['height','measurement']] = full_2019.height.str.split("ft",expand=True,)
```

```python
full_2019[['period','measurement2']] = full_2019.period.str.split("s",expand=True,)
```

```python
full_2019[['temp','measurement2']] = full_2019.temp.str.split("°c",expand=True,)
```

```python
full_2019.drop(["measurement","measurement2"],axis = 1,inplace= True)
```

```python
full_2019.rename(columns={"temp": "temp_°c"},inplace = True)
```

```python
full_2019.rename(columns={"height": "height_ft"},inplace = True)
```

```python
full_2019.rename(columns={"period": "period_sec"},inplace = True)
```

```python
full_2019 = full_2019.astype({"height_ft": float, "swell_angle": int,"period_sec": int, "wind_kmh": int, "wind_angle":int,"temp_°c":int  })
```

```python
def anio(x):
    return f"{x}/2019"

full_2019.date = full_2019.date.apply(anio)

full_2019["date"] = pd.to_datetime(full_2019.date)
```

```python
full_2019.to_csv("full_2019.csv", index=False)
```
