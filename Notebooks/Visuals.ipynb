### Analysis
- From the first two diagrams of the subplot, we can see that both northern hemisphere spots swell angles fluctuate around the 320 and 0 degree mark, both of these swell directions being north ( as 359 degrees and 1 degree are nearly the same direction due to it being measured roundly)
- We can also see from that both spots heighest swell heights are found in the NW to N range ( between 290 degrees and 10 degrees)

- For the two bottom diagrams, these being the spots found in the southern hemisphere, we can see that the dominant swell direction is quiet different from the ones above. The padang- padang graph is a clear example of how most southern hemisphere spots recieve their swell, usually rounding between 200-240 degree range. This being swells from SSW to WSW. On the other hand the Box´s behaviour is different. Even though most of the swell still also comes from the south, we can see that it pulls more towards the western direction then in padang-padang. This most probably happens due to the location of the box; it being just off the west coast of western australia. Here the swells are created in a similar way to those reaching bali, but due to the geographical features of where the spot is found, the southern formed swell swoops more westwards as it reachs the australian coast, probably due to the underwater unevenness creating a swell magnet

```python
def sustituye(x):
    for key,value in long_lat.items():
        if x == key:
            return value
```

```python
data_lat_long['Location'] = data_lat_long.Spot.apply(sustituye)
```

```python
def sustituye_lat(x):
    for key,value in long_lat.items():
        if x == key:
            return value[0]
        
```

```python
data_lat_long['Latitude'] = data_lat_long.Spot.apply(sustituye_lat)
```

```python
def sustituye_long(x):
    for key,value in long_lat.items():
        if x == key:
            return value[1]
```

```python
data_lat_long['Longitud'] = data_lat_long.Spot.apply(sustituye_long)
```
